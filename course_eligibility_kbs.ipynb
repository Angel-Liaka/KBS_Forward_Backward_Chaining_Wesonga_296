This notebook demonstrates a simple Knowledge Based System that uses **Forward Chaining** and **Backward Chaining** to determine course eligibility.

```python
student_name = "Angel Wesonga"
registration_number = "672296"

print("Student:", student_name)
print("Registration Number:", registration_number)
```

```python
class KnowledgeBase:
    
    def __init__(self):
        self.facts = set()
        self.rules = []

    def add_fact(self, fact):
        self.facts.add(fact)

    def add_rule(self, rule):
        self.rules.append(rule)

    def forward_chain(self):
        
        print("\nForward Chaining Process:")
        
        new_fact_added = True
        
        while new_fact_added:
            new_fact_added = False
            
            for rule in self.rules:
                if rule["conditions"].issubset(self.facts) and rule["result"] not in self.facts:
                    
                    print("Rule Fired:", rule["conditions"], "→", rule["result"])
                    
                    self.facts.add(rule["result"])
                    new_fact_added = True

        return self.facts
```

```python
rules = [

    {"conditions": {"Programming I", "Mathematics I"},
     "result": "Data Structures"},

    {"conditions": {"Data Structures", "Computer Organization"},
     "result": "Algorithms"},

    {"conditions": {"Programming I", "Statistics"},
     "result": "Machine Learning Basics"},

    {"conditions": {"Algorithms", "High GPA"},
     "result": "Advanced AI"},

    {"conditions": {"Programming I"},
     "result": "Intro to Software Development"}
]
```

```python
print("\n===== Student Test Case 1 =====")

student_courses = {"Programming I", "Statistics"}
gpa = 3.2

print("Student: Alice")
print("Completed Courses:", student_courses)
print("GPA:", gpa)

kb = KnowledgeBase()

for course in student_courses:
    kb.add_fact(course)

if gpa >= 3.0:
    kb.add_fact("High GPA")

for r in rules:
    kb.add_rule(r)

print("Initial Facts:", kb.facts)

final_facts = kb.forward_chain()

print("\nFinal Facts:", final_facts)
```

```python
def backward_chaining(goal, facts, rules):
    
    print("\nChecking Goal:", goal)
    
    if goal in facts:
        print("Goal already known.")
        return True
    
    for rule in rules:
        
        if rule["result"] == goal:
            
            print("Goal depends on:", rule["conditions"])
            
            for condition in rule["conditions"]:
                
                if condition not in facts:
                    print("Missing requirement:", condition)
                    return False
            
            return True
    
    return False
```

```python
print("\n===== Student Test Case 1 =====")

student_courses = {"Programming I", "Statistics"}
gpa = 3.2

print("Student: Alice")
print("Completed Courses:", student_courses)
print("GPA:", gpa)

kb = KnowledgeBase()

for course in student_courses:
    kb.add_fact(course)

if gpa >= 3.0:
    kb.add_fact("High GPA")

for r in rules:
    kb.add_rule(r)

print("Initial Facts:", kb.facts)

final_facts = kb.forward_chain()

print("\nFinal Facts:", final_facts)
```

```python
print("\nBackward Chaining Query")
print("Is the student eligible for Machine Learning Basics?")

goal = "Machine Learning Basics"

result = backward_chaining(goal, kb.facts, rules)

if result:
    print("Result: PROVED")
else:
    print("Result: NOT PROVED")
```
